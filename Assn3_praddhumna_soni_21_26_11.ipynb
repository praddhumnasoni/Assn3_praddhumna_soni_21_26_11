6. Replace the NaN and String entries in Num_Bath with previous entry

```python
df['NUM_BATH'] = df['NUM_BATH'].replace('HURLEY', 'NaN') #to replace string to Nan string
df['NUM_BATH'] = df['NUM_BATH'].replace('NaN', np.nan) # to replace Nan to integer
df['NUM_BATH'].fillna(method='ffill', inplace=True) # Replace the NaN with previous entry using forward fill
df
```

7.Replace the empty values in “SQ_FT” with the mean of the all entries.

```python
df['SQ_FT'] = df['SQ_FT'].replace('--', 'NaN') #to replace string to Nan string
df['SQ_FT'] = df['SQ_FT'].replace('NaN', np.nan) # to replace Nan to integer
df['SQ_FT'] = pd.to_numeric(df['SQ_FT'])
df['SQ_FT'].fillna(value=df['SQ_FT'].mean(), inplace=True)
df
```
